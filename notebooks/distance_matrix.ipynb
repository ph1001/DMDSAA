# Distance Matrix

```python
# imports
from sklearn import datasets
```

```python
# abstract
dataset = datasets.load_iris()
```

```python
# dictionary
dataset.keys()
```

```python
dataset["feature_names"]
```

```python
data = dataset["data"]
data  # data is a numpy array data structure. Think of it as a matrix of data (or as an excel spreadsheet)
```

```python
data.shape
```

```python
# euclidean distance of 2 observations

point1 = data[50] # this is the 51st observation
point2 = data[100] # this is the 101st observation

sum((point2-point1)**2)**(1/2)

# We forgot: Variables have different scales!!!
```

```python
# initialize distance matrix. What will be its final shape?
dist = []
```

```python
# Build the distance matrix. Use 2 for loops, the append list method and the euclidean distance formula      
rowCount = data.shape[0]
print(rowCount)

for i in range(rowCount):
    p1 = data[i]
    dist_row = []
    for j in range(rowCount):
        p2 = data[j]
        dist_row.append(sum((p2-p1)**2)**(1/2))
    # append to distance matrix
    dist.append(dist_row)
```

```python
dist
```

```python
sns.heatmap(dist)
```
